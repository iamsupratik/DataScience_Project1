##12. Join users to transactions, keeping all rows from transactions and only matching rows
from users (left join)

```python
result = pd.merge(transactions,
                 users[['UserID', 'User', 'Gender', 'Registered', 'Cancelled']],
                 on='UserID', 
                 how='left')
result['Registered'].fillna(pd.NaT, inplace=True)
result['Cancelled'].fillna(pd.NaT, inplace=True)
result
```

##14. Join users to transactions, keeping only rows from transactions and users that match
via UserID (inner join)

```python
result = pd.merge(transactions,
                 users[['UserID', 'User', 'Gender', 'Registered', 'Cancelled']],
                 on='UserID', 
                 how='inner')
result['UserID'] = result['UserID'].astype(int)
result['Cancelled'].fillna(pd.NaT, inplace=True)
result
```

##15. Join users to transactions, displaying all matching rows AND all non-matching rows
(full outer join)

```python
result = pd.merge(transactions,
                 users[['UserID', 'User', 'Gender', 'Registered', 'Cancelled']], 
                 how='outer')
result.loc[:, ['TransactionDate','Registered','Cancelled']] = result.loc[:, ['TransactionDate','Registered','Cancelled']].fillna(pd.NaT)
result
```

```python
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```
